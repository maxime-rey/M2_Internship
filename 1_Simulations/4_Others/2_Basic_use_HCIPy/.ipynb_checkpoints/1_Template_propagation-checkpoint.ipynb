# PSF simulation with HCIPy

## **Info**

In pupil_grid.x, values are distance in meters from the center, increases when going &uparrow;, &rightarrow;

## **Initialisation**

```python
%matplotlib inline
```

```python
import hcipy as hci
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = (15, 9)
plt.rcParams['image.origin'] = True
```

## **Fraunhofer propagation**

```python
%%time

diam_aperture = 1                             
wvlgth = 1                                           
focal_length = 1 

grid_size_pix = 1024                                         # [pix] Nb of pixels on the grid (ex: 1024x1024)
grid_diam = 2*diam_aperture                                # [m] Grid on which field is sampled

pupil_grid = hci.make_pupil_grid(grid_size_pix, grid_diam)        
aperture = hci.circular_aperture(diam_aperture)(pupil_grid)  # Create aperture (field), mets des 0 et des 1 sur grid


sampl_focal = 32                                            #échantillonage q (8samples per λ/D) 
fov = 16                                                    # [λ/D], radius
                                                            #Why si q diminue, nb rings increases in "power" ?
focal_grid = hci.make_focal_grid(pupil_grid, sampl_focal, fov)
prop = hci.FraunhoferPropagator(pupil_grid, focal_grid, wvlgth, focal_length)
wf = hci.Wavefront(aperture, wvlgth)
img = prop.forward(wf) 

print('Fresnel number = {}'.format(diam_aperture**2/(wvlgth*focal_length)))
print('Fraunhofer distance = {}'.format(2*diam_aperture**2/wvlgth))

plt.subplot(1, 3, 1)
hci.imshow_field(np.log10(img.intensity / img.intensity.max()), vmin=-5)
plt.colorbar(fraction=0.046, pad=0.04)
plt.xlabel('x [λ/D]')
plt.ylabel('y [λ/D]')
plt.title('log(I) normalised')

plt.subplot(1, 3, 2)
hci.imshow_field(np.log10(img.power), vmin=-5)
plt.colorbar(fraction=0.046, pad=0.04)
plt.xlabel('x [λ/D]')
plt.ylabel('y [λ/D]')
plt.title('power: I*(weight at pix)')  

plt.subplot(1, 3, 3)
hci.imshow_field(img.phase, vmin=-5)
plt.colorbar(fraction=0.046, pad=0.04)
plt.xlabel('x [λ/D]')
plt.ylabel('y [λ/D]')
plt.title('Phase')                        #Weight if grid not uniform

plt.tight_layout()
plt.show()
```

## **Fresnel propagation**

```python
diam_aperture_2 = 0.01     
wvlgth_2 = 500e-9             
focal_length_2 = 2 
distance = focal_length_2

grid_size_pix_2 = 1024                                         # [pix] Nb of pixels on the grid (ex: 1024x1024)
grid_diam_2 = 1.5*diam_aperture_2                                # [m] Grid on which field is sampled

pupil_grid_2 = hci.make_pupil_grid(grid_size_pix_2, grid_diam_2)        
aperture_2 = hci.circular_aperture(diam_aperture_2)(pupil_grid_2)  # Create aperture (field), mets des 0 et des 1 sur grid

fresnel_prop = hci.FresnelPropagator(pupil_grid_2, distance)
wf_2 = hci.Wavefront(aperture_2, wvlgth_2)
img_2 = fresnel_prop(wf_2)

plt.subplot(1, 3, 1)
hci.imshow_field(np.log10(img_2.intensity / img_2.intensity.max()), vmin=-5)
plt.colorbar(fraction=0.046, pad=0.04)
plt.xlabel('x [λ/D]')
plt.ylabel('y [λ/D]')
plt.title('Normalised intensity {}'.format(wvlgth_2)) 


plt.subplot(1, 3, 2)
hci.imshow_field(img_2.intensity, vmin=0)  #.power: I*(weight at pix), weigth if grid not uniform
plt.colorbar(fraction=0.046, pad=0.04)
plt.xlabel('x [λ/D]')
plt.ylabel('y [λ/D]')
plt.title('Intensity')

plt.subplot(1, 3, 3)
hci.imshow_field(img_2.phase)
plt.colorbar(fraction=0.046, pad=0.04)
plt.xlabel('x [λ/D]')
plt.ylabel('y [λ/D]')
plt.title('Phase')

plt.tight_layout()
plt.show()
```

## **Plot grid and aperture**

```python
plt.subplot(2, 2, 1)
plt.plot(pupil_grid.x, pupil_grid.y, '+', label='pupil grid unit')
plt.xlabel('x [m]')
plt.ylabel('y [m]')
plt.title('Fraunhofer pupil grid')
plt.axis('equal')
#plt.legend()

plt.subplot(2, 2, 2)
hci.imshow_field(wf.intensity,label='wavefront')             #can also get .phase .amplitude
plt.colorbar(fraction=0.046, pad=0.04)
plt.xlabel('x [m]')
plt.ylabel('y [m]')
plt.title('Init. wavefront intensity')

plt.subplot(2, 2, 3)
plt.plot(pupil_grid_2.x, pupil_grid_2.y, '+', label='pupil grid unit')
plt.xlabel('x [m]')
plt.ylabel('y [m]')
plt.title('Fresnel pupil grid')
plt.axis('equal')
#plt.legend()

plt.subplot(2, 2, 4)
hci.imshow_field(wf_2.intensity,label='wavefront')           #can also get .phase .amplitude
plt.colorbar(fraction=0.046, pad=0.04)
plt.xlabel('x [m]')
plt.ylabel('y [m]')
plt.title('Init. wavefront intensity')

plt.tight_layout()
plt.show()
```
