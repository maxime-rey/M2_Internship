# PSF simulation with HCIPy

## **Initialisation**

```python
%matplotlib inline
```

```python
import hcipy as hci
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = (15, 9)
plt.rcParams['image.origin'] = True
```

## **Parameters**

```python
diam_aperture = 1                                            # [m] Diam telescope
wvlgth = 1                                                   # [m]
focal_length = 1                                             # [m]

grid_diam = 1.5*diam_aperture                                # [m] Grid on which field is sampled


print('Fresnel number = {}'.format(diam_aperture**2/(wvlgth*focal_length)))
print('Fraunhofer distance = {}'.format(2*diam_aperture**2/wvlgth))
```

## **Fraunhofer propagation**

```python
grid_size_pix = 1024                                         # [pix] Nb of pixels on the grid (ex: 1024x1024)

sampl_focal = 8                                              #échantillonage q (8samples per λ/D) 
fov = 16                                                     # [λ/D], radius

pupil_grid = hci.make_pupil_grid(grid_size_pix, grid_diam)        
aperture = hci.circular_aperture(diam_aperture)(pupil_grid)  # Create aperture (field)

                                   #Why si echantillonage diminue, nb rings increases i, power ?
focal_grid = hci.make_focal_grid(pupil_grid, sampl_focal, fov)

""" Do a wf list: for each lambda (wf is monochromatic)"""
"""
bandwidth = 0.2

focal_total = 0
for wlen in np.linspace(1 - bandwidth / 2., 1 + bandwidth / 2., 11):
    wavefront = Wavefront(telescope_pupil, wlen)
    focal_total += prop(wavefront).intensity
    
imshow_field(np.log10(focal_total / focal_total.max()), vmin=-5)
"""
wf = hci.Wavefront(aperture, wvlgth)                         # Create wavefront with Lambda=1

prop = hci.FraunhoferPropagator(pupil_grid, focal_grid, wvlgth, focal_length)
```

## **Fresnel propagation**

```python
diam_aperture_2 = 0.01                        # Gotta remove this afterwards to unify w/ fraun
wvlgth_2 = 660e-9  
grid_diam_2 = 1.5*diam_aperture_2                             #Change it to 2, 8 -> get funny things
                                                             # At >8 looks realistic
print('Fresnel number = {}'.format(diam_aperture_2**2/(wvlgth_2*focal_length)))

pupil_grid_2 = hci.make_pupil_grid(grid_size_pix, grid_diam_2)
aperture_2 = hci.circular_aperture(diam_aperture_2)(pupil_grid_2)

""" Do a wf list: for each lambda (wf is monochromatic)"""
wf_2 = hci.Wavefront(aperture_2, wvlgth_2)                   # Create wavefront with Lambda=1

fresnel_prop = hci.FresnelPropagator(pupil_grid_2, focal_length)
```

## **Propagate through nothing**

```python
img = prop.forward(wf) 
img_2 = fresnel_prop(wf_2)
```

## **Plot Fraunhofer after propagation**

```python
plt.subplot(1, 3, 1)
hci.imshow_field(np.log10(img.intensity / img.intensity.max()), vmin=-5)
plt.colorbar(fraction=0.046, pad=0.04)
plt.xlabel('x [λ/D]')
plt.ylabel('y [λ/D]')
plt.title('log(I) normalised')

plt.subplot(1, 3, 2)
hci.imshow_field(np.log10(img.power), vmin=-5)
plt.colorbar(fraction=0.046, pad=0.04)
plt.xlabel('x [λ/D]')
plt.ylabel('y [λ/D]')
plt.title('power: I*(weight at pix)')  

plt.subplot(1, 3, 3)
hci.imshow_field(img.power, vmin=-5)
plt.colorbar(fraction=0.046, pad=0.04)
plt.xlabel('x [λ/D]')
plt.ylabel('y [λ/D]')
plt.title('Phase')                        #Weight if grid not uniform

plt.tight_layout()
plt.show()
```

## **Plot Fresnel after propagation**

```python
plt.subplot(1, 3, 1)
hci.imshow_field(np.log10(img_2.intensity / img_2.intensity.max()), vmin=-5)
plt.colorbar(fraction=0.046, pad=0.04)
plt.xlabel('x [λ/D]')
plt.ylabel('y [λ/D]')
plt.title('Normalised intensity {}'.format(wvlgth)) 

plt.subplot(1, 3, 2)
hci.imshow_field(img_2.intensity)
plt.colorbar(fraction=0.046, pad=0.04)
plt.xlabel('x [λ/D]')
plt.ylabel('y [λ/D]')
plt.title('Intensity')

plt.subplot(1, 3, 3)
hci.imshow_field(img_2.phase)
plt.colorbar(fraction=0.046, pad=0.04)
plt.xlabel('x [λ/D]')
plt.ylabel('y [λ/D]')
plt.title('Phase')

plt.tight_layout()
plt.show()
```

## **Aberration**

```python
aberration = hci.SurfaceAberration(pupil_grid, 1/4, 1)       #peak-to-valley abrration & diam where its computed

img = prop(aberration(wf))

hci.imshow_field(np.log10(img.intensity / img.intensity.max()), vmin=-5)
plt.colorbar()
plt.show()
```

## **Detector**

```python
flat_field = 0.01
dark = 10

detector = hci.NoisyDetector(focal_grid, dark_current_rate=dark, flat_field=flat_field)
wf.total_power = 1000
img = prop(aberration(wf))

detector.integrate(img, 0.5)                                 # Integration time
image = detector.read_out()

hci.imshow_field(np.log10(image), vmax=np.log10(image).max(), vmin=0)
plt.colorbar()
plt.show()
```

## **Atmosphere**

```python
%%time
D_tel = 8.2 # meter
wavelength = 1e-6 # meter

pupil_grid = hci.make_pupil_grid(512, D_tel)
focal_grid = hci.make_focal_grid(pupil_grid, 8, 16, wavelength=wavelength)
prop = hci.FraunhoferPropagator(pupil_grid, focal_grid)

aperture = hci.circular_aperture(D_tel)(pupil_grid)
```

```python
%%time
fried_parameter = 0.2                                        # [m]
outer_scale = 20                                             # [m], "L0"
velocity = 10                                                # [m/s]
height = 10000                                               # [m]

Cn_squared = hci.Cn_squared_from_fried_parameter(fried_parameter, 900e-9)
layer = hci.InfiniteAtmosphericLayer(pupil_grid, Cn_squared, outer_scale, velocity, height)
```

```python
%%time
wf = hci.Wavefront(aperture, wavelength)
img = prop(layer(wf))

plt.subplot(1, 2, 1)
hci.imshow_field(layer.phase_for(wavelength), cmap='RdBu')
plt.colorbar(fraction=0.046, pad=0.04)
plt.xlabel('x ?')
plt.ylabel('y ?')
plt.title('Atmosphere')

plt.subplot(1, 2, 2)
hci.imshow_field(np.log10(img.intensity / img.intensity.max()), vmin=-3)
plt.colorbar(fraction=0.046, pad=0.04)
plt.xlabel('x ?')
plt.ylabel('y ?')
plt.title('Light through atm.')

plt.tight_layout()
plt.show()
```

```python
%%time                                                       
                                                             # Dizaine de secondes
pupil_grid3 = hci.make_pupil_grid(256, 1.5)
focal_grid3 = hci.make_focal_grid(pupil_grid3, 2, 64)
prop = hci.FraunhoferPropagator(pupil_grid3, focal_grid3)

pup = hci.circular_aperture(1)(pupil_grid3)

# Make multi-layer atmosphere
layers3 = hci.make_standard_atmospheric_layers(pupil_grid3, outer_scale)

atmos = hci.MultiLayerAtmosphere(layers3, scintilation=True)
# Set the seeing parameters and reset
atmos.Cn_squared = hci.Cn_squared_from_fried_parameter(0.1, 550e-9)
atmos.reset()

# Use super-Gaussian to avoid edge effects
p = np.exp(-(pupil_grid3.as_('polar').r/0.68)**20)
wf3 = atmos(hci.Wavefront(hci.Field(p, pupil_grid3), 2e-6))

# Show scintillation field
hci.imshow_field(wf3.intensity * pup)
plt.show()
```
